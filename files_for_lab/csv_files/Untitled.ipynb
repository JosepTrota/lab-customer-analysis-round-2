```python
from IPython.display import Image
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

data=pd.read_csv('marketing_customer_analysis.csv')
```

```python
data.info()
```

```python
data.describe()
```

```python
freq = {}
for item in data['Response']:
    if item in ['Yes', 'No']:
        if (item in freq):
            freq[item] += 1
        else:
            freq[item] = 1
            
height = list(freq.values())
bars = list(freq.keys())
y_pos = np.arange(len(bars))

plt.bar(y_pos, height)

plt.xticks(y_pos, bars)

plt.show()

freq
```

```python
data.groupby('Sales Channel')['Response'].value_counts().unstack().plot.bar(ylabel='Responses')
```
